```python
import numpy as np
```

```python
# создание массива из списка
list_of_nums = [1,2,3]
a = np.array(list_of_nums)
print(type(a))
print(a.shape)
```

```python
# создание массивов с помощью встроенных функций
a = np.ones((4,4))
b = np.eye(4)
print(a)
print(b)
print(a + b)
```

```python
# создание массива последовательности и форматированный вывод
np.set_printoptions(precision = 1)
print(np.arange(0, 1, 0.01))
```

```python
a = np.empty((3,2,4))
print(f"Форма массива {a.shape}")
print(f"Тип данных массива {a.dtype}")
```

```python
a = np.arange(0, 10, 1)
print(f"a = {a}")
print(f"Индекс 0 = {a[0]}, индекс 9 = {a[9]}")
print(f"Срез со 2 по 4 эелемент включительно {a[1:5]}")
```

```python
a = np.eye(3)
b = np.ones((3,3))
print(f"Поэлементное сложение \n {a + b}")
print(f"Поэлементное умножение \n {a * b}")
```

```python
a = np.array([1,1,1,2,3,4])
print(f"Среднее арифметическое {np.mean(a)}")
print(f"Медиана {np.median(a)}")
```

```python
# Пример с оценками
grades = np.array([[4,5,4],[3,4,5],[5,4,3]])
titles = np.array(["Математика", "Литература", "ОБЖ"])
print("Среднее по предментам:")
for i in range(titles.size):
    print(f"{titles[i]}: {np.mean(grades[i]):.2f}") 

print(f"Среднее по всем {np.mean(grades):.2f}")
```

```python
# Пример с температурой
week_temps = np.array([12.5, 14.0, 13.2, 15.1, 16.8, 18.0, 17.3])
print("Температуры по дням недели:", week_temps)
print("Минимальная температура:", np.min(week_temps))
print("Максимальная температура:", np.max(week_temps))
print(f"Средняя температура: {np.mean(week_temps):.2f}")
```

```python
# Пример с покупками
sales = np.array([[120, 80, 60], [150, 90, 70],[130, 110, 65]])
categories = np.array(["Еда", "Быт", "Одежда"])
print("Общие покупки по категориям:")
for i in range(categories.size):
    print(f"{categories[i]}: {np.sum(sales[:, i])} руб")
```
